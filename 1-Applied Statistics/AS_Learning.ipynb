In statistics and probability theory, the Bayes theorem (also known as the Bayes rule) is a mathematical formula used to determine the conditional probability of events. Essentially, the Bayes theorem describes the probability of an event based on prior knowledge of the conditions that might be relevant to the event.

$P(A|B) =  \frac{P(B|A)P(A)} {P(B)}$ 

Where:

P(A|B)  the probability of event A occurring, given event B has occurred

P(B|A)  the probability of event B occurring, given event A has occurred

P(A)  the probability of event A

P(B)  the probability of event B

Imagine you are a financial analyst at an investment bank. According to your research of publicly-traded companies, 60% of the companies that increased their share price by more than 5% in the last three years replaced their CEOs during the period.

At the same time, only 35% of the companies that did not increase their share price by more than 5% in the same period replaced their CEOs. Knowing that the probability that the stock prices grow by more than 5% is 4%, find the probability that the shares of a company that fires its CEO will increase by more than 5%.

Before finding the probabilities, you must first define the notation of the probabilities.

P(A)  the probability that the stock price increases by 5%

P(B)  the probability that the CEO is replaced

P(A|B)  the probability of the stock price increases by 5% given that the CEO has been replaced

P(B|A)  the probability of the CEO replacement given the stock price has increased by 5%.

Using the Bayes theorem, we can find the required probability:

$P(A|B) =  \frac{P(B|A)P(A)} {P(B|A-)P(A-) + P(B|A+)P(A+)}$ 

$P(A|B) =  \frac{0.60X0.04} {0.60X0.04 + 0.35X(1-0.04)}$ 

P(A|B) =  0.067 or 67%

Thus, the probability that the shares of a company that replaces its CEO will grow by more than 5% is 6.67%.

To make it uniform we will now remove these using lambda function and will retain only the first year (in combined year only) in a seperate column called "YearLaunch" and we create a new column called "Age" which contains the age of the team in years.


```python
def comb2single(x):
    if 'to' in x:
        return x.split('to')[0]
    elif '-' in x:
        return x.split('-')[0]
    if '~' in x:
        return x.split('~')[0]
    if '_' in x:
        return x.split('_')[0]
    else:
        return x
    
basketball_data["YearLaunch"] = basketball_data["TeamLaunch"].apply(comb2single) 
#convert YearLaunch to int64
basketball_data = basketball_data.astype({"YearLaunch":np.int64})
today = date.today()
#subtract yearlaunch to current date to get team age
basketball_data["Age"] = basketball_data["YearLaunch"].apply(lambda x: today.year-x)
```

### Measure of skewness
Skewness is a measure of extent to which a distribution differs from a normal distribution.

The rule of thumb I would use here:
* If the skewness is between -0.5 and 0.5, the data are **fairly symmetrical**
* If the skewness is between -1 and -0.5(negatively skewed) or between 0.5 and 1(positively skewed), the data are **moderately skewed**
* If the skewness is less than -1(negatively skewed) or greater than 1(positively skewed), the data are **highly skewed**

Determining skewness of `Score`,`TournamentChampion`,`Runner-up` columns and ploting the results after fitting a norm distribution

```python
# Measure of Skewness and kurtosis for `Score`,`Win_Percent` columns
print("Skewness of 'Score': {}\n".format(basketball_data['Score'].skew().round(3)))

print("Skewness of 'Win_Percent': {}\n".format(basketball_data['Win_Percent'].skew().round(3)))
```

### Outliers imputation
In statistics, an outlier is an observation point that is distant from other observations.

**Ways to detect outliers**
* ***Box Plot***: In descriptive statistics, a box plot is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending vertically from the boxes (whiskers) indicating variability outside the upper and lower quartiles, hence the terms box-and-whisker plot and box-and-whisker diagram. Outliers may be plotted as individual points.

* ***Scatter Plot***: A scatter plot , is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. The data are displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.

* ***Z-score***: The Z-score is the signed number of standard deviations by which the value of an observation or data point is above the mean value of what is being observed or measured. In most of the cases a threshold of 3 or -3 is used i.e if the Z-score value is greater than or less than 3 or -3 respectively, that data point will be identified as outliers.

* ***IQR score***: The interquartile range (IQR), also called the midspread or middle 50%, or technically H-spread, is a measure of statistical dispersion, being equal to the difference between 75th and 25th percentiles, or between upper and lower quartiles, IQR = Q3  Q1. In other words, the IQR is the first quartile subtracted from the third quartile; these quartiles can be clearly seen on a box plot on the data. It is a measure of the dispersion similar to standard deviation or variance, but is much more robust against outliers.

I would be using Boxplot (to visualize), IQR (to display the columns) for outliers

#### B. Write the null hypothesis and alternative hypothesis. [1 Mark]

In testing whether there are any significant difference between Funds raised by companies that are still operating vs companies that closed down, the null hypothesis states that difference between the funds raised by companies that are operating and companies that are closed is not significant,  $\mu{O}$ equals $\mu{C}$. The 
alternative hypthesis states that the Funds raised by companies is different for operating and closed, $\mu{O}$ $\neq$ $\mu{C}$.
* $H_0$: $\mu{O}$ - $\mu{C}$ =      0
* $H_A$: $\mu{O}$ - $\mu{C}$ $\neq$  0

#### C. Test for significance and conclusion [1 Mark]

Here we select $\alpha$ = 0.05 and sample size < 30 and population standard deviation is not known.

```python
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest
operating = mydata[mydata['OperatingState']=='Operating']['Funds_in_million']
closed = mydata[mydata['OperatingState']=='Closed']['Funds_in_million']
z_stat,p_val = ztest(operating,closed)

print('Z-statistic: ', '%1.3f' % z_stat, '\n\nP value: ',  '%1.3f' % p_val)

print('\nResult of Statistical Test:')

if p_val > 0.05:
    print('Since', '%.3f' % p_val, 'is greater than the significance level(0.05), Failed to Reject Null Hypothesis (H0) - Difference between the funds raised by companies that are operating and companies that are closed are not significant')
else:
    print('Since', '%.3f' % p_val, 'is less than the significance level(0.05), Null Hypothesis (H0) rejected - Funds raised by companies is different for operating and closed')
```

```python
from scipy.stats import ttest_ind

t_stats, p_value = ttest_ind(operating,closed)
print('T-statistic: ', '%1.3f' % t_stats, '\n\nP value: ',  '%1.3f' % p_value)

print('\nResult of Statistical Test:')

if p_value > 0.05:
    print('Since', '%.3f' % p_val, 'is greater than the significance level(0.05), Failed to Reject Null Hypothesis (H0) - Difference between the funds raised by companies that are operating and companies that are closed are not significant')
else:
    print('Since', '%.3f' % p_val, 'is less than the significance level(0.05), Null Hypothesis (H0) rejected - Funds raised by companies is different for operating and closed')
```

#### G. Write your hypothesis comparing the proportion of companies that are operating between winners and contestants: [2 Mark]

It is observed that the proportions are different but to determine if they are statistically significantly different, we will do a test of proportions via the z-test or Chi sq test. Again, we can perform this test as the sample size is > 30

**Null Hypothesis (H0) - Proportion of companies that are operating across winners and contestants are similar**

**Alternate Hypothesis (H1) - Proportion of companies that are operating across winners and contestants are different.**

#### H. Test for significance and conclusion [1 Mark]

### Z-Test

```python
from statsmodels.stats.proportion import proportions_ztest

z_stat,p_val = proportions_ztest([n_w_o,n_c_o],[n_W,n_C])

print('Z-statistic: ', '%1.3f' % z_stat, '\n\nP value: ',  '%1.3f' % p_val)

print('\nResult of Statistical Test:')

if p_val > 0.05:
    print('Since', '%.3f' % p_val, 'is greater than the significance level(0.05), Failed to Reject Null Hypothesis (H0) - Proportion of companies that are operating across winners and contestants are similar')
else:
    print('Since', '%.3f' % p_val, 'is less than the significance level(0.05), Null Hypothesis (H0) rejected - Proportion of companies that are operating across winners and contestants are different.')
```
